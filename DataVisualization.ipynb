```python
from wordcloud.wordcloud import WordCloud
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

df = pd.read_csv('/content/preprocess_dataset.csv')
stopwords = set(STOPWORDS)


wc = WordCloud(background_color='white',max_words=3000,stopwords=stopwords)
wc.generate(str(df))

fig = plt.figure()
fig.set_figwidth(12)
fig.set_figheight(6)

plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()

counts = WordCloud().process_text(str(df))
# print(dict(sorted(counts.items(),reverse=True)))
datacount = list(sorted(counts.items(),reverse=True))
data = pd.DataFrame(datacount).reset_index()
data.sort_values(by=1, ascending=False)
```
